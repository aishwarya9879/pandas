```python

# Going to plot these on 3 axes within the same figure. Totally fine if you did them
# on individual axes or just with the seaborn function, of course.

import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(sat_f.Rate, bins=10, hist=True, kde=False, color='gold' )
```

```python
fig, axarr = plt.subplots(1, 3, figsize=(18,7))
axarr[0] = sns.distplot(sat_f.Rate, bins=10,ax=axarr[0], kde=False, color='gold' )
axarr[1] = sns.distplot(sat_f.Math, bins=10,ax=axarr[1], kde=False, color='red' )
axarr[2] = sns.distplot(sat_f.Verbal, bins=10,ax=axarr[2], kde=False, color='blue' )

axarr[0].set_title('Rate', fontsize=20)
axarr[1].set_title('Math', fontsize=20)
axarr[2].set_title('Verbal', fontsize=20)

plt.show()
```
